Data-Pre Processing

Encoding the Categorical features

```python
#encoding sex column
insurance_dataset.replace({"sex": {'male':0, 'female':1}}, inplace=True)
#encoding smoker column
insurance_dataset.replace({"smoker": {'yes':0, 'no':1}}, inplace=True)
#encoding region column
insurance_dataset.replace({"region": {'southeast':0, 'southwest':1, 'northeast': 2, 'northwest': 3}}, inplace=True)
58
```

Splitting the features and target

```python
X = insurance_dataset.drop(columns="charges", axis=1)
Y = insurance_dataset['charges']
```

```python
print(X)
print(Y)
```

Splitting the data into Training data & Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state =2)
##test size 0.2 means 20% for testing data and 80% for training
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Linear Regression

```python
## loading the linear regressino model
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

Model Evaluation

```python
## prediction on training data
training_data_prediction = regressor.predict(X_train)
```

```python
# R square value 0-1, if close to 1 -> performing well
r2_train = r2_score(Y_train, training_data_prediction)
print('R squared value: ', r2_train)
```

```python
# prediction on test_data
testing_data_prediction = regressor.predict(X_test)
```

```python
##checking the R-square value
r2_test = r2_score(Y_test, testing_data_prediction)
print("R squared value of r2_score: ",r2_test)
```

Creating a Predictive System

```python
def encode_input(*input_data):
    # Define the encoding mappings
    sex_mapping = {'male': 0, 'female': 1}
    smoker_mapping = {'yes': 0, 'no': 1}
    region_mapping = {'southeast': 0, 'southwest': 1, 'northeast': 2, 'northwest': 3}

    # Unpack the input data
    age, sex, bmi, children, smoker, region = input_data

    # Encode the categorical variables
    sex_encoded = sex_mapping.get(sex, sex)
    smoker_encoded = smoker_mapping.get(smoker, smoker)
    region_encoded = region_mapping.get(region, region)
    return (age, sex_encoded, bmi, children, smoker_encoded, region_encoded)


input_data = encode_input(31, "male", 25.74, 0, "no", "southeast")

# chaning input data into numpy array
input_data_as_array = np.asarray(input_data)

#reshaping the array
input_data_reshaped = input_data_as_array.reshape(1,-1)
prediction = regressor.predict(input_data_reshaped)
print("The insurance cost is USD: ", prediction[0])
```
